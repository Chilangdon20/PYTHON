### Una funci√≥n de aplanamiento de tweets.

Normalmente estamos interesados ‚Äã‚Äãen cientos o miles de tweets. Para este prop√≥sito, tiene sentido definir una funci√≥n para aplanar el archivo JSON lleno de tweets.

Llamemos a esta funci√≥n `flatten_tweets()`. Usaremos esta funci√≥n varias veces en este curso y la cambiaremos ligeramente a medida que manejamos diferentes tipos de datos.



```
import json
def flatten_tweets(tweets_json):
    """Aplana los diccionarios
     de tweets para que JSON relevante 
     est√© en un diccionario de nivel superior."""
    tweets_list = []
    
    # Iterar cada tweet
    for tweet in tweets_json:
        tweet_obj = json.loads(tweet)
    
        # Almacena el nombre de pantalla del usuario en 'user-screen_name'
        tweet_obj['user-screen_name'] = tweet_obj['user']['screen_name']
    
        # Comprueba si se trata de un tweet de m√°s de 140 caracteres
        if 'extended_tweet' in tweet_obj:
            # Almacene el texto del tweet extendido en 'extended_tweet-full_text'
            tweet_obj['extended_tweet-full_text'] = tweet_obj['extended_tweet']['full_text']
    
        if 'retweeted_status' in tweet_obj:
            # Almacena el nombre de pantalla del usuario de retweet en 'retweeted_status-user-screen_name'
            tweet_obj['retweeted_status-user-screen_name'] = tweet_obj['retweeted_status']['user']['screen_name']

            # Almacena el texto del retweet en 'retweeted_status-text'
            tweet_obj['retweeted_status-text'] = tweet_obj['retweeted_status']['text']
            
        tweets_list.append(tweet_obj)
    return tweets_list
```
