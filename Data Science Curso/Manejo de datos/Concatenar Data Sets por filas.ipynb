# Concatenar Data Sets por filas

Algunos de los estudios de datos más interesantes provienen de la combinación de diferentes fuentes de datos. Estas operaciones pueden involucrar cualquier cosa, desde una concatenación muy directa de dos conjuntos de datos diferentes, hasta combinaciones y combinaciones de estilo de base de datos más complicadas que manejan correctamente cualquier superposición entre los conjuntos de datos. Series y DataFrames se crean teniendo en cuenta este tipo de operaciones, y Pandas incluye funciones y métodos que hacen que este tipo de datos se discuta de manera rápida y sencilla.

## En python hay 2 tipos de ejes:
    * axis = 0 denota el eje horizontal.
    * axis = 1 denota el eje vertical.

```python
#Con la funcion concat unimos los dos data set
data_vino = pd.concat([vino_rojo,vino_blanco], axis = 0)
```

```python
data_vino.shape
```

```python
#APILAR = Podemos observar que importa el orden en el cual se concatenan los datos en este caso los primero datos que se muestran son los del vino rojo
data_vino.head()
```

```python
#Extraemos 10 filas del primer data set y del segundo y las juntamos
data1 = data_vino.head(10)
data2 = data_vino[300:310]
data3 = data_vino.tail(10)
```

```python
vino_scramble = pd.concat([data1,data2,data3], axis = 0)
```
