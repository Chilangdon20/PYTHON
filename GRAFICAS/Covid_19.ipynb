```
import pandas as pd 
df = pd.read_csv('totalescv.csv')
df=df.T
df.head() 
df= df.drop(['date'])
df=df.reset_index()
df.head()
#Cambiamos los nombres de las collmnas para que sea mas facil trabajar con ellos
df=df.rename({'index':'Paises',53:'total'},axis=1)
df.head()
```

```
#Instalamos la ibreria pycountry que hara la conversion de paises a su abreviatura
pip install pycountry
```

```
#Creamos una funcion 
import pycountry
def get_alpha_3(location):
  try:
    return pycountry.countries.get(name=location).alpha_3
  except:
    return None
df['Code'] = df['Paises'].apply(lambda x: get_alpha_3(x))    
df.head()
```

```
#GRAFICA
import plotly.express as px
fig = px.choropleth(df,locations='Code',
                    color='total',hover_name='Paises',
                    color_continuous_scale=px.colors.sequential.Plasma)
fig.show()                    

```
